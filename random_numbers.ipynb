```
import tensorflow as tf
import numpy as np
import math
import matplotlib.pyplot as plt
```

```
sess = tf.Session()

def p(func):
  print(sess.run(func))
```

```
p(tf.zeros([2,3]))
p(tf.zeros([2,3],dtype=tf.int32))
p(tf.ones([2,3],dtype=tf.int32))
p(tf.ones([2,4],dtype=tf.float32))
```

```
p(tf.zeros_like(tf.ones([2,4])))
p(tf.fill([2,3],2))
p(tf.fill([2,3],2.0))
```

```
linspace_0 = tf.linspace(1.,10,8)

plt.plot(sess.run(linspace_0),'ro')
plt.show()

p(tf.range(0,5))
tf.range(0,10,3)

rand_n = tf.random_normal([2,3],mean=9)
plt.plot(sess.run(rand_n))
plt.show()
```

```
tf.random_uniform([2,3])

uniform0 = tf.random_uniform([2,3])
plt.plot(sess.run(uniform0))
plt.show()

uniform1 = tf.random_uniform([2,3],minval=5)
plt.plot(sess.run(uniform1))
plt.show()

uniform2 = tf.random_uniform([2,3],maxval=15)
plt.plot(sess.run(uniform0))
plt.show()
```

```
x = [*range(-10,11)]
y = [*map(lambda g:1/(1+math.exp(-g)),x)]
plt.plot(x,y,'ro')
plt.show()
```
